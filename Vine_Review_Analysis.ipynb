```python
import os 
import pandas as pd 
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
#check df
df.sample(10)
```

```python
df.info()
```

```python
#The data is filtered to create a DataFrame or table where there are 20 or more total votes (5 pt)
df_greater_20 = df[df['total_votes']>= 20]
df_greater_20
```

```python
#number of helpful_votes divided by total_votes is equal to or greater than 50%

df_helful_greater = df[df['helpful_votes'] / df['total_votes'] >= .50]
df_helful_greater
```

```python
# retrieves all the rows where a review was written as part of the Vine program (paid)
df_vine_review = df[df['vine'] == 'Y']

df_vine_review
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid)
df_vine_noreview = df[df['vine'] == 'N']

df_vine_noreview
```

```python
total_reviews = df['review_id'].count()
total = f'Total reviews = {total_reviews}'
print(total)
```

```python
total_fs_reviews = df[df['star_rating']==5].count()
total_fs_reviews = pd.DataFrame(total_fs_reviews).transpose()
total_fs_reviews = total_fs_reviews['star_rating'].sum()
total_fs = f'5-Star reviews = {total_fs_reviews}'
print(total_fs)
```

```python
paid_FSR = df[(df['vine'] =='Y') & (df['star_rating']==5)].count()
paid_FSR_sum = paid_FSR['star_rating'].sum()
FS_precentage = paid_FSR_sum / total_fs_reviews *100
total_paid_fs = f'% of 5-Star reviews paid = {FS_precentage}%'
print(total_paid_fs)
```

```python
unpaid_FSR = df[(df['vine'] =='N') & (df['star_rating']==5)].count()
unpaid_FSR_sum = unpaid_FSR['star_rating'].sum()
FS_precentage = unpaid_FSR_sum / total_fs_reviews *100
untotal_paid_fs = f'% of 5-Star reviews unpaid = {FS_precentage}%'
print(untotal_paid_fs)
```
